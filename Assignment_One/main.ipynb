## Here is the link to the dataset: FIFA 21 messy, raw dataset for cleaning/ exploring (kaggle.com)

## Perform the following exploration and cleaning on the dataset:

1. Convert the height and weight columns to numerical forms

2. Remove the unnecessary newline characters from all columns that have them.

3. Based on the 'Joined' column, check which players have been playing at a club for more than 10 years!

4. 'Value', 'Wage' and "Release Clause' are string columns. Convert them to numbers. For eg, "M" in value column is Million, so multiply the row values by 1,000,000, etc.

5. Some columns have 'star' characters. Strip those columns of these stars and make the columns numerical

6. Which players are highly valuable but still underpaid (on low wages)? (hint: scatter plot between wage and value)

```python
import pandas as pd
```

```python
#  read the csv file
fifa_df = pd.read_csv('fifa2rawdatav2.csv')
fifa_df.head(20)
```

```python
fifa_df.info()
fifa_df.isnull()
```

```python
# convert height and weight to numerical form
def toNumeric(x):
    if 'cm' in x:
        num = x.split('c')[0]
    else: 
        num = x.split('k')[0]
    return num

fifa_df['Height'] = fifa_df['Height'].apply(toNumeric)
fifa_df['Weight'] = fifa_df['Weight'].apply(toNumeric)
#  rename the column to show unit
fifa_df.rename(columns={'Height': 'Height(cm)', 'Weight':'Weight(kg)'}, inplace=True)
# remove any non numeric values and converting to int
fifa_df['Weight(kg)'] = fifa_df['Weight(kg)'].str.replace(r'\D', '', regex=True).astype(int)
fifa_df['Height(cm)'] = fifa_df['Height(cm)'].str.replace(r'\D', '', regex=True).astype(int)

pd.DataFrame(fifa_df['Weight(kg)'].head(5))
```

```python
# remove newline characters
fifa_df = fifa_df.map(lambda x: x.replace('\n', '') if isinstance(x, str) else x)
fifa_df.select_dtypes(include='object')
```

```python
# player who have been playing for more that 10 years
from datetime import datetime

# convert 'Joined' column to datetime
fifa_df['Joined'] = pd.to_datetime(fifa_df['Joined'], format='%b %d, %Y')

# # filter players who have been playing for more than 10 years
ten_years_ago = datetime.today() - pd.DateOffset(years=10)
long_term_players = fifa_df[fifa_df['Joined'] < ten_years_ago]

long_term_players[['Name', 'Club', 'Joined']]
```

```python
# 'Value', 'Wage' and "Release Clause' convert them to numbers

# function to replace the convert the short symbol value to numbers
def toActualValue(x):
    syb = None
    if 'M' in x:
        syb = 'M'
    elif 'K' in x:
        syb = 'K'  
    x = float(x.replace('€', '').split(syb)[0])
    if syb == 'M':
        x = x * 1000000
        return x
    elif syb == "K":
        x = x * 1000
        return x
    else: return x


# print(toActualValue('€120'))
fifa_df['Value'] = fifa_df['Value'].apply(toActualValue).astype(float) 
fifa_df['Wage'] = fifa_df['Wage'].apply(toActualValue).astype(float)
fifa_df['Release Clause'] = fifa_df['Release Clause'].apply(toActualValue).astype(float)
```

```python
fifa_df.select_dtypes(include='object')
```

```python
# remove start chars from columns and make the numerical
# function to strip the star char
def RemoveStarChar(x):
    x = x.split('★')[0]
    return x

fifa_df['W/F'] = fifa_df['W/F'].apply(RemoveStarChar).astype(int)
fifa_df['SM'] = fifa_df['SM'].apply(RemoveStarChar).astype(int)
fifa_df['IR'] = fifa_df['IR'].apply(RemoveStarChar).astype(int)
pd.DataFrame(fifa_df['IR'])
```

```python
# replace all NaN values with 0
fifa_df.fillna(0, inplace=True)
```
