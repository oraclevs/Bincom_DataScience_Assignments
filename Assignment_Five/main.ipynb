## Pipe lines for all my Models

```python
from sklearn.pipeline import Pipeline
from sklearn.discriminant_analysis import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import FunctionTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn import set_config
set_config(display='diagram') # configuration for viewing the diagram structure of the pipeline
```

```python
#  define transform functions (placeholders)
def to_numeric(x):
    return x


def to_actual_value(x):
    return x


def remove_star_char(x):
    return x
```

```python
# fifa dataset model pipline

fifa_pipeline = Pipeline([
    ('to_numeric', FunctionTransformer(to_numeric)),
    ('to_actual_value', FunctionTransformer(to_actual_value)),
    ('remove_star_char', FunctionTransformer(remove_star_char)),
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler()),
    ('regressor', LinearRegression())
])
fifa_pipeline
```

```python
# boston house price prediction model pipeline
housing_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler()),
    ('regressor', LinearRegression())
])
housing_pipeline
```

```python
#  titanic survival prediction al model  pipeline 
titanic_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('classifier', RandomForestClassifier(n_estimators=300, random_state=42))
])
titanic_pipeline
```

```python
# spam mail prediction model dataset 

spam_pipeline = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('classifier', LogisticRegression())
])
spam_pipeline
```
